## Exists

The `exists` function is used to check if a category exists in the taxonomy. It returns `True` if the category exists, `False` otherwise.

```python
#| export

def exists(category: str) -> bool:
    """Check if a category, or a subcategory exists in the arXiv taxonomy."""
    if category in main or category in sub or category == 'all':
        return True
    else:
        return False
```
